```python
#function
dx=lambda t,x: 0.65*(1-x/5.4)*x
actual= lambda t,x: 5.4/(1+((5.4/0.5)-1)*np.exp(-0.65*t))

h=0.5; ti=0; tf=30
t=0;x=0.5

n=int((tf-ti)/h)

data_bigstepsize = [(0,0.5)]
data_actual=[0.5]
xvals_1=[0.5] #inital condition

#Euler's Method
for i in range (1, n+1):
    x=x+dx(t,x)*h
    t=t+h
    data_bigstepsize.append((t,x))
    xvals_1.append(x)
    data_sol=actual(t,x)
    data_actual.append(data_sol)
    
#Append to A Table
df = pd.DataFrame(data_bigstepsize, columns=['t','x'])
df['actual']=data_actual
error1=abs(np.array(data_actual)-np.array(xvals_1))
df['error']=error1

df
```

```python
def f(x):
    return 0.65*(1-x/8.1)*x-(1.2*x**2 / (1 + x**2))

a = -1  # lower bound of interval
b = 1  # upper bound of interval
tolerance = 1e-6

while (b-a)/2 > tolerance:
    c = (a+b)/2  # midpoint of interval
    if f(c) == 0:
        break
    elif f(a)*f(c) < 0:
        b = c
    else:
        a = c

print("Zero found at x = {:.6f}".format(c))
```

```python
# Define the derivative function
def f(t,x):
    return 0.65*(1-(x/8.1))*x-((1.2*x**2 )/ (1 + x**2))

# Create a grid of x and y values
t = np.linspace(0, 30, 31)
x = np.linspace(0, 8, 31)
T, X = np.meshgrid(t, x)

# Compute the derivative at each point
dxdt=f(T,X)

#Normalize Vectors
U=(1/(1+dxdt**2)**.5)*np.ones(T.shape)
V=(1/(1+dxdt**2)**.5)*dxdt
   
# Plot the vector field
plt.figure(figsize = (15, 11))
plt.quiver(T, X, U, V, color='b',pivot='mid')
plt.plot(df_harvest1['t'],df_harvest1['x'], label="Euler's Method Graph, Intial Condition 84")
plt.plot(df_harvest2['t'],df_harvest2['x'],label="Euler's Method Graph, Intial Condition 24")
plt.plot(df_harvest3['t'],df_harvest3['x'],label="Euler's Method Graph, Intial Condition 18")
plt.plot(df_harvest4['t'],df_harvest4['x'], label="Euler's Method Graph, Intial Condition 6")

plt.legend()
plt.title("Direction Field and Inital Conditons for Deer Population")
plt.xlabel("Time in Years")
plt.ylabel("Population of Deer in Dozens")
plt.grid()
plt.show()




```
